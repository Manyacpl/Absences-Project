<h4>Import library and file</h4>

```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv("Absenteeism-data.csv")
raw_csv_data
```

```python
# I create a copy of the DataFrame as a backup for the original df

df = raw_csv_data.copy()
df
```

```python
# I check if the date contains missing data

df.info()
```

<h4>Drop ID </h4>

```python
# Usuwam kolumne ID. Nie jest mi potrzebna ponieważ przygotowuję analize dla całej populacji pracowników

df = df.drop(['ID'], axis = 1)
```

<h4>.get_dummies()</h4>

```python
# Replaces the 'Reason for Absence' column with dummy numbers

reason_columns = pd.get_dummies(df['Reason for Absence'], dtype=int)
```

```python
reason_columns
```

```python
# Using the new 'check' column, I verify that each row has only one reason for absence (more than one would indicate a data error)

reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns['check'].sum(axis=0)
```

```python
reason_columns['check'].unique()
```

```python
# After checking that the data contains no errors, I remove the 'check' column

reason_columns = reason_columns.drop(['check'], axis = 1)
reason_columns
```

<h4>Group the Reason for Absence</h4>

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
# Removing the "Reason for Absence" column to avoid multicollinearity

df =  df.drop(['Reason for Absence'], axis = 1)
```

```python
df
```

```python
# Grouping reasons for absence into appropriate classes. Group 1 is diseases, group 2 is childbirth,
# group 3 is poisoning, group 4 is minor reasons like dentist visits, medical appointments


reason_type_1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis = 1)
```

```python
reason_type_3.unique()
```

<h4>Concatenate Column Values</h4>

```python
df
```

```python
# Adding the 'reason' column to the DataFrame

df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df
```

```python
df.columns.values
```

```python
# Renaming the columns

column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
df.head()
```

<h4>Reorder Columns</h4>

```python
# Reorganizing the columns in the DataFrame so that 'Reasons' are at the beginning.

column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
```

```python
df.head()
```

<h4>Checkpoint</h4>

```python
df_checkpoint = df.copy()
df_checkpoint
```

<h4>Date</h4>

```python
df_checkpoint['Date']
```

```python
type(df_checkpoint['Date'][0])
```

```python
# Formatting the data in the 'Date' column to Timestamp format

df_checkpoint['Date'] = pd.to_datetime(df_checkpoint['Date'], format= '%d/%m/%Y')
```

```python
type(df_checkpoint['Date'][2])
```

<h4>Extract the Month Value</h4>

```python
df_checkpoint['Date'][0]
```

```python
df_checkpoint['Date'][0].month
```

```python
list_months = []
list_months
```

```python
# Adding months from df_checkpoint['Date'] to the empty list list_months using a loop

for i in range(700):
    list_months.append(df_checkpoint['Date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
# Adding the 'Month Value' column

df_checkpoint['Month Value'] = list_months
```

```python
df_checkpoint.head()
```

<h4>Extract the Day of the Week</h4>

```python
# Creating a column that shows the day of the absence (0 - Mon...6 - Sun)

df_checkpoint['Date'][500].weekday()
```

```python
#Tworzę funkcje date_to_weekday która będzie pobierać weekday z zadanego dnia

def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
# Creating the function date_to_weekday that will retrieve the weekday from a given date

df_checkpoint['Day of the Week'] = df_checkpoint['Date'].apply(date_to_weekday)
```

```python
df_checkpoint.head()
```

```python
# Removing the now unnecessary 'Date' column

df_checkpoint = df_checkpoint.drop(['Date'], axis = 1)
df_checkpoint
```

```python
# Changing the order of the columns

df_checkpoint.columns.values
```

```python
column_names_new = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_checkpoint = df_checkpoint[column_names_new]
df_checkpoint.head()
```

```python
df_checkpoint_2 = df_checkpoint.copy()
```
